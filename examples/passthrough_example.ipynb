### Explain the drivers of trends in WAC, WAL, SMM

1. Weighted Average Coupon (WAC):
   - Interest Rate Environment: The prevailing interest rates influence the mortgage rates for new loans. When interest rates rise, newer mortgages typically have higher interest rates, which increases the WAC of the MBS pool. Conversely, falling interest rates result in lower WAC as newer mortgages have lower interest rates.
   - Refinancing Activity: Changes in interest rates also affect borrower behavior regarding refinancing. Lower interest rates generally lead to higher refinancing activity, causing the WAC of the MBS pool to decrease as borrowers replace higher-interest mortgages with lower-interest ones.

2. Weighted Average Life (WAL):
   - Prepayment Speeds: Prepayment speeds are influenced by factors such as changes in interest rates, economic conditions, and borrower behavior. Higher prepayment speeds result in a shorter WAL, as the principal is paid down more quickly due to early repayments from refinancing or home sales.
   - Loan Characteristics: The types of mortgages in the pool, such as fixed-rate or adjustable-rate mortgages, and features like prepayment penalties or restrictions, can impact prepayment behavior and, consequently, the WAL of the MBS pool.

3. Single Monthly Mortality (SMM):
   - Interest Rate Environment: Changes in interest rates impact borrower decisions regarding refinancing. Lower interest rates generally lead to higher prepayment speeds and, therefore, higher SMM as borrowers refinance to secure more favorable rates.
   - Economic Conditions: Economic factors such as employment levels, housing market trends, and borrower sentiment can influence prepayment behavior. For example, an improving economy may lead to increased refinancing activity, resulting in higher SMM.
   - Loan Aging: As loans age, the likelihood of prepayments typically increases. Thus, the SMM may increase over time as loans in the pool mature and approach their natural prepayment periods.
