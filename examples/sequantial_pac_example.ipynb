```python
tranch1_wal_li = []
tranch2_wal_li = []
tranch3_wal_li = []
tranch4_wal_li = []
psa_arr = np.linspace(50, 500, 50)

for psa in psa_arr:
    wals_li = Sequential(psa).wals()
    tranch1_wal_li.append(wals_li[0])
    tranch2_wal_li.append(wals_li[1])
    tranch3_wal_li.append(wals_li[2])
    tranch4_wal_li.append(wals_li[3])

plt.plot(psa_arr, tranch1_wal_li, label='tranch1_WAL')
plt.plot(psa_arr, tranch2_wal_li, label='tranch2_WAL')
plt.plot(psa_arr, tranch3_wal_li, label='tranch3_WAL')
plt.plot(psa_arr, tranch4_wal_li, label='tranch4_WAL')
plt.xlabel('PSA')
plt.ylabel('WAL')
plt.legend()
plt.show()
```

### List the tranches in declining order of sensitivity (of WAL as a function of PSA)

Tranch 4 > Tranch 3 > Tranch2 > Tranch 1

For a Sequential structure, a tranch in back position is more sensitive to PSA since there are more precedent principals that influence its payment plan.

## PAC

```python
pac_100 = PAC(100)
pac_100.plot()
```

```python
for psa in np.linspace(50, 500, 10):
    PAC(psa).plot()
```

The PAC busts up if psa is below or above the psa band of (100, 250). The PAC does not bust up if psa is between the psa band of (100, 250). That is because the companion is fully filled up before maturity. It is notable that the bust-up happens near maturity if a psa is lower than the lower bound(100) of the psa band. When a psa gets higher than the upper bound (250) of the psa band, the bust-up happens earlier.
