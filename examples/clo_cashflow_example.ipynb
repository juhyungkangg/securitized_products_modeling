```python
# B Lost Default Rate
start = 0.0104807415
end   = 0.010480742
num = 1000

for default_rate in np.linspace(start, end, num):
    val = B_lose_point(default_rate)
    if (val > 0):
        B_lose_default_rate = default_rate
    if (val > 1 / 1000000):
        break
        
print(B_lose_default_rate)
```

```python
print("The rate is around", np.round(B_lose_default_rate*100, 5), "%.")
```

```python
losing_val = B_lose_point(B_lose_default_rate)*1000000

print("B loses $", losing_val,"under", np.round(B_lose_default_rate*100, 5), "% of default rate.")
```
